```python
sns.set(style='darkgrid')
rs = np.random.RandomState(0)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

sns.residplot(x,y,lowess=True,color='g');
```
