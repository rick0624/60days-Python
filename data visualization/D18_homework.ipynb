```python
import matplotlib.pyplot as plt
import numpy as np
```

# Q1

```python
x = np.arange(0,3*np.pi,0.01)
y_cos = np.cos(x)

plt.plot(x,y_cos)
plt.savefig('cos.png',dpi=300,format="png")
plt.show()
```

# Q2

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
```

```python
T = np.add(Y,X)
plt.scatter(X,Y,s=50,c=T,alpha=0.3)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
