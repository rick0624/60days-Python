```python
import matplotlib.pyplot as plt
import numpy as np
```

# Q1

```python
x = np.arange(0,3*np.pi,0.01)
y_cos = np.cos(x)

plt.plot(x,y_cos)
plt.savefig('cos.png',dpi=300,format="png")
plt.show()
```
